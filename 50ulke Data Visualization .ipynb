```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
df = pd.read_csv('50ulke.csv')
df.head()
```

```python
# The pairplot() char 
```

```python
sns.pairplot(df)
```

```python
# The barplot () Char Of Data Visualization
```

```python
sns.barplot(x="Continent", y="Per person", data=df, palette="viridis")
```

```python
sns.barplot(x="Continent", y="Growth", data=df, palette="viridis")
```

```python
sns.barplot(x="Continent", y="National Income", data=df, palette="viridis")
```

```python
#  The countplot() Char Data Visualization 
```

```python
sns.countplot(x="Continent", data=df, palette='viridis')
```

```python
 # The stripplot() char Data Visualization
```

```python
sns.stripplot(x="Continent", y="Growth", data=df, palette="viridis")
```

```python
sns.stripplot(x="Continent", y="National Income", data=df, palette="viridis")
```

```python
# the swarmplot() char Data Visualization 
```

```python
sns.swarmplot(x="Continent", y="Growth", data=df, palette="viridis")
```

```python
sns.swarmplot(x="Continent", y="Per person", data=df, palette="viridis")
```

```python
# lmplot char data visualization
```

```python
sns.lmplot(x="Per person", y="Growth", data=df, height=7, aspect=1.3)
```

```python
# The jointplot() Char Data Visualization
```
